```python
import matplotlib.pyplot as plt 
df = pd.read_csv('./smoke_detection_iot.csv')
X1 = df['PM2.5']
y1 = df['NC2.5']

fig = plt.figure(figsize=(12,6))
plt.subplot(121)
plt.scatter(X1, y1, s=1, c='b', marker = 's')
plt.xlabel('PM2.5')
plt.ylabel('NC2.5')
plt.legend()

plt.show()
```
